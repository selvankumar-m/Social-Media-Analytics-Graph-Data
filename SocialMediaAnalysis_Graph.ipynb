Step2: import NetworkX, pandas and numpy packages

```python
import networkx as nx
import pandas as pd
import numpy as np
```

Step3: Read smaese_data.csv file into pandas

```python
data = pd.read_csv('/content/smaese_data.csv')
```

Step4: Show top-15 rows of data frame

```python
data.head(15)
```

### PART-2

Step6: Use from_pandas_edgelist() method from networkx package and create airline data graph

```python
el = nx.Graph()
el = nx.from_pandas_edgelist(data, 'Origin','Dest', create_using = nx.DiGraph())
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10, 8))
nx.draw_shell(el, with_labels=True)
```

Step7: Show All nodes

```python
el.nodes()
```

Step8: How many cities in the route map?

```python
data.Origin.unique()
```

Step9: Show All edges

```python
el.edges()
```

### PART-3

Step11: Plot network graph using draw_networkx() method

```python
figure(figsize=(10,8))
nx.draw_networkx(el, with_labels=True)
```

### PART-4

Step13: Find shortest route between airports PBI and HOU based on AirTime by using shortest_path() method and print it

```python
Short = nx.shortest_path(el, source='PBI',target='HOU')
for path in Short:
  print(path)
```

Step15: Find the length of shortest route between airports PBI and HOU based on AirTime by using shortest_path_length() method and print it

```python
nx.shortest_path_length(el,source='PBI',target='HOU',weight='AirTime')
```
